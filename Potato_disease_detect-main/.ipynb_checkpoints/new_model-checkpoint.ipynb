```python
# converting images to array 
def convert_image_to_array(image_dir):
    try:
        image = cv2.imread(image_dir)
        if image is not None : 
            image = cv2.resize(image,(256,256))
            return img_to_array(image)
        else : 
            return np.array([])
    except Exception as e: 
        print (f"Error :{e}")
        return None
```

```python
dir = "C:\\Users\\jagje\\OneDrive\\Desktop\\new plant\\PE_dataset\\PlantVillage"
image_list , label_list = [] , [] 
all_labels = ['Potato___Early_blight' , 'Potato___Late_blight' ]
binary_labels = [0 , 1]
temp = -1 

# reading and converting image to numpyarray 
for directory in ['Potato___Early_blight' , 'Potato___Late_blight' ]:
    plant_image_list = listdir(f"{dir}/{directory}")
    temp += 1
    for files in plant_image_list: 
        image_path = f"{dir}/{directory}/{files}"
        image_list.append(convert_image_to_array(image_path))
        label_list.append(binary_labels[temp])
        
```
